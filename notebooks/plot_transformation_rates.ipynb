```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
expt = 'rho_hconst050'
regions = ['natl', 'npac', 'so', 'globe']
d = dict()

for r in regions:
    d[r] = np.load('../data/%s_%s.npz' % (expt, r))
T = np.arange(d['natl']['A'].shape[0])/12.

# create seasonal and annual means
A_seasonal = dict()
A_annmean = dict()
month = np.mod(T*12,12)+1
seasons = dict(
    djf = (month==12) | (month==1) | (month==2),
    mam = (month==3) | (month==4) | (month==5),
    jja = (month==6) | (month==7) | (month==8),
    son = (month==9) | (month==10) | (month==11) )
for r in regions:
    A_annmean[r] = d[r]['A'].mean(axis=0)
    A_seasonal[r] = dict()
    for s, sidx in seasons.iteritems():
        A_seasonal[r][s] = d[r]['A'][sidx].mean(axis=0)
```

```python
r = 'globe'
Sv = 1e6

f = plt.figure(figsize=(12,3.5))

plt.subplot(131)
plt.plot(d[r]['rholevs'][1:] - 1000., -A_seasonal[r]['djf'].sum(axis=0)/ Sv, 'k-', linewidth=2)
plt.plot(d[r]['rholevs'][1:] - 1000., -A_seasonal[r]['djf'].T/ Sv)
plt.grid()
plt.legend(['Net','HF','FWF','Mixing', 'Cabbeling'], loc='lower left')
plt.title('DJF')

plt.subplot(132)
plt.plot(d[r]['rholevs'][1:] - 1000., -A_annmean[r].sum(axis=0)/ Sv, 'k-', linewidth=2)
plt.plot(d[r]['rholevs'][1:] - 1000., -A_annmean[r].T/ Sv)
plt.grid()
plt.title('Annual Mean')

plt.subplot(133)
plt.plot(d[r]['rholevs'][1:] - 1000., -A_seasonal[r]['jja'].sum(axis=0)/ Sv, 'k-', linewidth=2)
plt.plot(d[r]['rholevs'][1:] - 1000., -A_seasonal[r]['jja'].T/ Sv)
plt.grid()
plt.title('JJA')
```

```python
plt.figure(figsize=(18,6))
plt.pcolormesh(T, d['so']['rholevs'][1:], np.cumsum(d['so']['A'][:,2,:], axis=1).T)
plt.colorbar()
```

```python
r = 'so'
plt.plot(d[r]['rholevs'][1:] - 1000., np.cumsum(A_annmean[r][2]))
for s in ['jja','djf']:
    plt.plot(d[r]['rholevs'][1:] - 1000., np.cumsum(A_seasonal[r][s][2]))
```

```python
expt = 'hconst050'
regions = ['natl', 'npac', 'so', 'globe']
dts = dict()

for r in regions:
    dts[r] = np.load('../data/A_TS_%s_%s.npz' % (expt, r))
T = np.arange(dts['natl']['A'].shape[0])/12.
```

```python
from watermasstools import transformation
# define TS space
NT = 100
NS = 98
Tlevs = np.linspace(-4,34,NT)
Slevs = np.linspace(28,41,NS)
globe = transformation.TSWaterMassRegion(Tlevs=Tlevs, Slevs=Slevs)
```

```python
#plt.pcolormesh(globe.S, globe.T, dts['globe']['A'].mean(axis=0)[1,0]/Sv, cmap='RdBu_r')
r = 'globe'
u = dts[r]['A'].mean(axis=0)[1,1]/Sv
v = dts[r]['A'].mean(axis=0)[0,1]/Sv
plt.figure(figsize=(10,10))
plt.quiver(globe.Sc, globe.Tc, u, v, scale=1000)
plt.xlim([32,38])
plt.clim([-20,20])
```
