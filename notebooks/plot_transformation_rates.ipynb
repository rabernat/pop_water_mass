# Surface Water Mass Transformation in Hi-Res POP Simulations #

This notebook shows preliminary results for the water mass transformation analysis of the ``` hybrid_v5_rel04_BC5_ne120_t12_pop62``` simulations. The analysis is based on the daily surface data. The terms calculated include the standard heat- and freshwater-flux-driven transformation but also the terms due to lateral (biharmonic) mixing and cabbelling. Lacking any depth dependence in the fields, I simply assume a surface layer depth of 50 m, over which the mixing and cabbelling are assumed to be coherent. (Using monthly 3D fields is not an option, because the time-averaging smooths them way too much.)

````python
import numpy as np
from matplotlib import pyplot as plt
import brewer2mpl
%matplotlib inline
````

## Load pre-computed data for plotting ##

````python
expt = 'WMT_rho_hconst050_new'
regions = ['natl', 'npac', 'so', 'globe']
regnames = ['North Atlantic', 'North Pacific', 'Southern Ocean', 'Global']
d = dict()

for r in regions:
    d[r] = np.load('../data/%s_%s.npz' % (expt, r))
T = np.arange(d['natl']['A'].shape[0])/12.

# create seasonal and annual means
A_seasonal = dict()
A_annmean = dict()
month = np.mod(T*12,12)+1
seasons = dict(
    djf = (month==12) | (month==1) | (month==2),
    mam = (month==3) | (month==4) | (month==5),
    jja = (month==6) | (month==7) | (month==8),
    son = (month==9) | (month==10) | (month==11) )
for r in regions:
    A_annmean[r] = d[r]['A'].mean(axis=0)
    A_seasonal[r] = dict()
    for s, sidx in seasons.iteritems():
        A_seasonal[r][s] = d[r]['A'][sidx].mean(axis=0)
````

## Plot Annual Mean Transformation Rates ##

````python
r = 'so'
Sv = 1e6

bmap = brewer2mpl.get_map('Set1', 'Qualitative', 4)
plt.rcParams['axes.color_cycle'] = bmap.mpl_colors
f = plt.figure(figsize=(12,9))

for n, r in enumerate(regions):

    plt.subplot(2,2,n+1)
    plt.plot(d[r]['rholevs'][1:] - 1000., -A_annmean[r].sum(axis=0)/ Sv, 'k-', linewidth=4)
    plt.plot(d[r]['rholevs'][1:] - 1000., -A_annmean[r].T/ Sv, linewidth=2)
    plt.xlabel(r'$\sigma_2$')
    plt.ylabel('transformation rate (Sv)')
    plt.grid()
    if n==0:
        plt.legend(['Net','HF','FWF','Mixing', 'Cabbeling'], loc='upper left')
    plt.title('Annual Mean Transformation : %s' % regnames[n])
plt.tight_layout()
````
