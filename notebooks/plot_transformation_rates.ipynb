```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
regions = ['natl', 'npac', 'so']
d = dict()
for r in regions:
    d[r] = np.load('../data/A_%s.npz' % r)
T = np.arange(d['natl']['A'].shape[0])/12.
Sv = 1e6
```

```python
r = 'npac'
plt.plot(d[r]['rholevs'][1:] - 1000., d[r]['A'].mean(axis=0).T/ Sv)
```

```python
plt.figure(figsize=(18,6))
plt.pcolormesh(T, d['natl']['rholevs'][1:], d['natl']['A'][:,2,:].T)
plt.colorbar()
```
