```python
def vector_divergence(pop, u, v):
    dxu = pop.nc.variables['DXU'][:]
    dyu = pop.nc.variables['DYU'][:]
    workx = 0.5 * u * dyu
    worky = 0.5 * v * dxu
    work1 = workx + np.roll(workx, 1, axis=-1)
    work1 = work1 - np.roll(work1, 1, axis=-2)
    work2 = worky + np.roll(worky, 1, axis=-2)
    work2 = work2 - np.roll(work2, 1, axis=-1)
    return work1 + work2
    
```
