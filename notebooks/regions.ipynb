```python
rho0 = p.rho[0]
natl = transformation.WaterMassRegion('natl',
                    basin_names=['Atlantic Ocean', 'GIN Seas'], latmin=15)
natl.initialize_mask(p)
natl.calculate_rholevs(rho0, rhomin=1022, rhomax=1028.5, nlevs=120, linear=True)
```

```python
A = dict()
for season in seasons:
    print season
    p = pop_model.POPFile(ddir + 'climatology.%s.nc' % season)
    p.initialize_gradient_operator()
    # calculate density
    rho0 = p.rho[0]
    # calculate heat and salt related density fluxes
    F_temp, F_salt = p.dens_flux[0]
    
    # T and S tendencies due to mixing
    bt = p.biharmonic_tendency(p.nc.variables['SST'][0])
    bs = p.biharmonic_tendency(p.nc.variables['SSS'][0])
    H_ml = p.nc.variables['HMXL_2'][0]/100.
    # transformation rate due to mixing
    F_Tmix = np.ma.masked_invalid( -p.alpha[0] * rho0 * H_ml * bt )
    F_Smix = np.ma.masked_invalid( p.beta[0] * rho0 * H_ml * bs )
    # for some reason, there are huge values going int F_Smix
    F_Smix = np.ma.masked_array( F_Smix, np.abs(F_Smix)>100.)
    
    print F_temp.max(), F_salt.max(), F_Tmix.max(), F_Smix.max()
    
    A[season] = natl.calculate_transformation_rate(rho0, F_temp, F_salt, F_Tmix, F_Smix )
    
A = np.array(A.values())    
```

```python
r = natl.rholevs[1:]-1000
Sv = 1e6
plt.plot(r, A.mean(axis=0)[0]/Sv, 'k--',
         r, A.mean(axis=0)[1]/Sv, 'k:',
         r, A.mean(axis=0)[2]/Sv, 'r-',         
         r, A.mean(axis=0)[3]/Sv, 'b-',         
         r, A.mean(axis=0).sum(axis=0)/Sv, 'k-', linewidth=2)
#plt.plot(r, A[0,0]/Sv, 'b--',
#         r, A[0,1]/Sv, 'b:',
#         r, A[0].sum(axis=0)/Sv, 'b-', linewidth=2)
#plt.plot(r, A[2,0]/Sv, 'r--',
#         r, A[2,1]/Sv, 'r:',
#         r, A[2].sum(axis=0)/Sv, 'r-', linewidth=2)
plt.grid()         
```
