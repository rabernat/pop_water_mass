```python
rho0 = p.rho[0]
natl = transformation.WaterMassRegion('natl',
                    basin_names=['Atlantic Ocean', 'GIN Seas'], latmin=15)
natl.initialize_mask(p)
natl.calculate_rholevs(rho0, rhomin=1022, rhomax=1028.5, nlevs=120, linear=True)
```

```python
A = dict()
for season in seasons:
    p = pop_model.POPFile(ddir + 'climatology.%s.nc' % season)
    # calculate density
    rho0 = p.rho[0]
    # calculate heat and salt related density fluxes
    F_temp, F_salt = p.dens_flux[0]
    A[season] = natl.calculate_transformation_rate(rho0, F_temp, F_salt )
A = np.array(A.values())    
```

```python
r = natl.rholevs[1:]-1000
Sv = 1e6
plt.plot(r, A.mean(axis=0)[0]/Sv, 'k--',
         r, A.mean(axis=0)[1]/Sv, 'k:',
         r, A.mean(axis=0).sum(axis=0)/Sv, 'k-', linewidth=2)
plt.plot(r, A[0,0]/Sv, 'b--',
         r, A[0,1]/Sv, 'b:',
         r, A[0].sum(axis=0)/Sv, 'b-', linewidth=2)
plt.plot(r, A[2,0]/Sv, 'r--',
         r, A[2,1]/Sv, 'r:',
         r, A[2].sum(axis=0)/Sv, 'r-', linewidth=2)
plt.grid()         
```
