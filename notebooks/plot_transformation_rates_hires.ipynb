```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
regions = ['natl', 'npac', 'so', 'globe']
expts = ['lores', 'hires']
A_seasonal = dict()
A_annmean = dict()
rholevs = dict()
d = dict()
for expt in expts:
    A_seasonal[expt] = dict()
    A_annmean[expt] = dict()
    rholevs[expt] = dict()
    for r in regions:
        d[r] = np.load('../data/A_rho_%s_%s.npz' % (expt, r))
        rholevs[expt][r] = d[r]['rholevs']
    T = np.arange(d['natl']['A'].shape[0])/12.

    # create seasonal and annual means
    month = np.mod(T*12,12)+1
    seasons = dict(
        djf = (month==12) | (month==1) | (month==2),
        mam = (month==3) | (month==4) | (month==5),
        jja = (month==6) | (month==7) | (month==8),
        son = (month==9) | (month==10) | (month==11) )
    for r in regions:
        # exclude mixing
        A_annmean[expt][r] = d[r]['A'][:,:2,:].mean(axis=0)
        A_seasonal[expt][r] = dict()
        for s, sidx in seasons.iteritems():
            A_seasonal[expt][r][s] = d[r]['A'][:,:2,:][sidx].mean(axis=0)
```

```python
r = 'so'
Sv = 1e6

f = plt.figure(figsize=(12,3.5))

plt.plot(rholevs['lores'][r][1:] - 1000., -A_annmean['hires'][r].sum(axis=0)/ Sv, 'k-', linewidth=2)
plt.plot(rholevs['hires'][r][1:] - 1000., -A_annmean['hires'][r][0]/ Sv, 'r-')
plt.plot(rholevs['hires'][r][1:] - 1000., -A_annmean['hires'][r][1]/ Sv, 'b-')
plt.plot(rholevs['hires'][r][1:] - 1000., -A_annmean['lores'][r].sum(axis=0)/ Sv, 'k--', linewidth=2)
plt.plot(rholevs['hires'][r][1:] - 1000., -A_annmean['lores'][r][0]/ Sv, 'r--')
plt.plot(rholevs['hires'][r][1:] - 1000., -A_annmean['lores'][r][1]/ Sv, 'b--')
plt.grid()
plt.legend(['Net (hires)','HF (hires)','FWF (hires)',
            'Net (lores)','HF (lores)','FWF (lores)'], loc='lower left')
plt.xlabel(r'$\sigma_2$')
plt.ylabel('Sv')
plt.title('Annual Mean Water Mass Transformation Rate South of 30S')

```
