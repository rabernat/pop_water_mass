```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

expts = ['lores_ctrl', 'hires_ctrl', 'lores_CO2', 'hires_CO2']
params = 'sigma2_hconst050'
Sv = 1e6
```

# Temporal Variability in Transformation Rates #

First let's just look at the temporal variability in annual mean transformation rate in the various experiments.
We take the monthly data, average over each year, and plot as a function of time and sigma.

```python
r = 'so'
fig = plt.figure(figsize=(18,9))
for n, expt in enumerate(['lores_ctrl', 'lores_CO2',  'hires_ctrl', 'hires_CO2']):
    plt.subplot(2,2,n+1)
    a = np.load('../data/WMT_%s_%s_%s.npz' % (expt, params, r))
    A, rholevs = a['A'], a['rholevs']-1000.
    Nt = A.shape[0]
    Tm = np.arange(Nt)
    Aann = np.add.reduceat(A, range(0,Nt,12), axis=0)/12.
    Ty = np.arange(Nt/12)
    Alevs = np.arange(-30,30,5)+2.5
    plt.contourf(Ty, rholevs[1:], -Aann[:,:2].sum(axis=1).T/Sv,
                 Alevs, cmap='RdBu_r', extend='both')
    plt.xlabel('time (years)')
    plt.ylabel(r'$\sigma_2$')
    plt.title(expt)
```

The most striking thing about this to me is the _complete lack of trend_ in the CO2 forcing experiments. Did I analyze the right data? On the other hand, there to appear to be significant differences in the _mean_ transformation rates between the control and CO2 experiments. I find this really puzzling.

# Differences in Mean Trasnformation Rates #

Motivated by the above figure, I didn't think it was meaningful to try to look for a "trend" in the CO2 forcing experiments. Instead, I just plot the overall mean for each set of experiments and their difference.

```python
regions = ['natl', 'npac', 'so', 'globe']


A_trend = dict()
A_mean = dict()
rholevs = dict()
d = dict()
for expt in expts:
    A_trend[expt] = dict()
    A_mean[expt] = dict()
    rholevs[expt] = dict()
    for r in regions:
        d[r] = np.load('../data/WMT_%s_%s_%s.npz' % (expt, params, r))
        rholevs[expt][r] = d[r]['rholevs']
    T = np.arange(d['natl']['A'].shape[0])/12.

    for r in regions:
        # exclude mixing
        A_mean[expt][r] = d[r]['A'][:,:2].mean(axis=0)
        A_trend[expt][r] = ( d[r]['A'][-60:,:2].mean(axis=0)
                           - d[r]['A'][:60,:2].mean(axis=0) )
```

```python
r = 'so'
Sv = 1e6

f = plt.figure(figsize=(12,5.5))

for expt, ls in zip(['hires_ctrl', 'lores_ctrl'], ['-', '--']):
    sigma = rholevs[expt][r][1:] - 1000.
    plt.plot(sigma, -A_mean[expt][r].sum(axis=0)/ Sv, 'k'+ls, linewidth=2)
    plt.plot(sigma, -A_mean[expt][r][0]/ Sv, 'r'+ls)
    plt.plot(sigma, -A_mean[expt][r][1]/ Sv, 'b'+ls)
plt.grid()
plt.legend(['Net (hires)','HF (hires)','FWF (hires)',
            'Net (lores)','HF (lores)','FWF (lores)'], loc='lower left')
plt.xlabel(r'$\sigma_2$')
plt.ylabel('Sv')
plt.title('Annual Mean Water Mass Transformation Rate South of 30S')

f = plt.figure(figsize=(12,5.5))

for expt, ls in zip(['hires_CO2', 'lores_CO2'], ['-', '--']):
    sigma = rholevs[expt][r][1:] - 1000.
    plt.plot(sigma, -A_mean[expt][r].sum(axis=0)/ Sv, 'k'+ls, linewidth=2)
    plt.plot(sigma, -A_mean[expt][r][0]/ Sv, 'r'+ls)
    plt.plot(sigma, -A_mean[expt][r][1]/ Sv, 'b'+ls)
plt.grid()
plt.legend(['Net (hires)','HF (hires)','FWF (hires)',
            'Net (lores)','HF (lores)','FWF (lores)'], loc='lower left')

plt.xlabel(r'$\sigma_2$')
plt.ylabel('Sv')
plt.title('Annual Mean Water Mass Transformation Rate with CO2 forcing')

f = plt.figure(figsize=(12,5.5))

for expt, ls in zip(['hires', 'lores'], ['-', '--']):
    sigma = rholevs[expt + '_CO2'][r][1:] - 1000.
    plt.plot(sigma, -(A_mean[expt + '_CO2'][r].sum(axis=0) -
                      A_mean[expt + '_ctrl'][r].sum(axis=0))/ Sv, 'k'+ls, linewidth=2)
    plt.plot(sigma, -(A_mean[expt + '_CO2'][r][0] - 
                      A_mean[expt + '_ctrl'][r][0])/ Sv, 'r'+ls)
    plt.plot(sigma, -(A_mean[expt + '_CO2'][r][1] -
                      A_mean[expt + '_ctrl'][r][1])/ Sv, 'b'+ls)
plt.grid()
plt.legend(['Net (hires)','HF (hires)','FWF (hires)',
            'Net (lores)','HF (lores)','FWF (lores)'], loc='lower left')

plt.xlabel(r'$\sigma_2$')
plt.ylabel('Sv')
plt.title('Difference (CO2 - ctrl)')
```

The difference figure is a bit noisy, but both models show:
* Decreased AABW formation with CO2 forcing
* Not much change in the upper CDW cell transformation
* Strong changes in the mode and intermediate water densities
