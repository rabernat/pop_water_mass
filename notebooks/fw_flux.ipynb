```python
import netCDF4
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import os

plt.rcParams['figure.figsize'] = (16,8)
plt.rcParams['image.origin'] = 'bottom'
```

```python
ddir = '/glade/p/ncgd0001/hybrid_v5_rel04_BC5_ne120_t12_pop62'
```

```python
fname = os.path.join(ddir, 'ice-hist', 'ann_avg_0067-0086.nc')
nc = netCDF4.Dataset(fname)
```

```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
```

```python
def extract_so(nc, varname, jlim=1000):
    """Returns lon, lat, data suitable for a basemap polar stereographic plot"""

    jlim = 1000

    try:
        lon = nc.variables['TLON'][...,:jlim,:]
        lat = nc.variables['TLAT'][...,:jlim,:]
    except KeyError:
        lon = np.ma.masked_equal(nc.variables['TLONG'][...,:jlim,:],0.)
        lat = np.ma.masked_equal(nc.variables['TLAT'][...,:jlim,:],0.)
        
    data = nc.variables[varname][...,:jlim,:]

    # this is a total hack to fetch regular lat / lon grids
    lon1d = lon[420]
    lat1d = lat[:,3000]

    i0 = np.argmin(lon1d)

    assert not np.any(lon1d.mask)
    assert not np.any(lat1d.mask)

    lon2d, lat2d = np.meshgrid(lon1d, lat1d)

    lon2d_fixed = np.roll(lon2d, -i0)
    lon1d_fixed = np.roll(lon1d, -i0)
    data_fixed = np.roll(data, -i0)
    
    return lon2d_fixed, lat2d, data_fixed
```

```python
def southern_ocean_pcolormesh(lon, lat, data, boundinglat=-50, ax=None, **kwargs):
    """Plot something in the southern ocean."""
    
    m = Basemap(projection='spstere',boundinglat=boundinglat,lon_0=180, ax=ax)
    #m.drawcoastlines()
    #m.fillcontinents(color='coral',lake_color='aqua')
    # draw parallels and meridians.
    m.drawparallels(np.arange(-80.,81.,20.))
    m.drawmeridians(np.arange(-180.,181.,30.), labels=[1,1,0,1])
    m.drawmapboundary(fill_color='0.5')
    #x, y = m(nc.variables['TLON'][:], nc.variables['TLAT'][:])
    return m.pcolormesh(lon, lat, data, latlon=True, **kwargs)
```

```python
lon, lat, data = extract_so(nc, 'fresh_ai')

fig = plt.figure()
ax = fig.add_subplot(111)
pc = southern_ocean_pcolormesh(lon, lat, data[0], ax=ax, cmap='PiYG')
pc.set_clim([-1,1])
ax.set_title("Annual Mean")
#plt.clim([-1,1])
plt.show()
```

```python
fig = plt.figure(figsize=(15,16))

season_names = ['jfm', 'amj', 'jas', 'ond']

for n, season in enumerate(season_names):
    
    fname = os.path.join(ddir, 'ice-hist', '%s_avg_0067-0086.nc' % season)
    ncs = netCDF4.Dataset(fname)
    
    lon, lat, data = extract_so(ncs, 'fresh_ai')
    data *= 365./100. # gives m/year
    
    ax = fig.add_subplot(2,2,n+1)
    pc = southern_ocean_pcolormesh(lon, lat, data[0], ax=ax, cmap='PiYG')
    pc.set_clim([-3,3])
    ax.set_title(season.upper())

cax = fig.add_axes([0.25,0.05,0.5,0.03])
cb = plt.colorbar(pc, cax=cax, orientation='horizontal')
cb.set_label('Freshwater Flux (m/year)')
plt.show()
```
