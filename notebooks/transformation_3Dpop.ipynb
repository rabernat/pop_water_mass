```python
rho_dummy = np.ones_like(p.mask)

natl = transformation.WaterMassRegion(
                    basin_names=['Atlantic Ocean', 'GIN Seas', 'Labrador Sea'], latmin=15)
natl.initialize_mask(p)
natl.calculate_rholevs(rho_dummy, rhomin=1022, rhomax=1028.5, nlevs=120, linear=True)
 
npac = transformation.WaterMassRegion(
                    basin_names=['Pacific Ocean'], latmin=15)
npac.initialize_mask(p)
npac.calculate_rholevs(rho_dummy, rhomin=1020, rhomax=1027, nlevs=120, linear=True)
 
so = transformation.WaterMassRegion(
                    basin_names=['Southern Ocean'], latmax=30)
so.initialize_mask(p)
so.calculate_rholevs(rho_dummy, rhomin=1022, rhomax=1028.5, nlevs=120, linear=True)

globe = transformation.WaterMassRegion()
globe.initialize_mask(p)
globe.calculate_rholevs(rho_dummy, rhomin=1020, rhomax=1030, nlevs=120, linear=True)

regions = {'North Atlantic': natl, 'North Pacific': npac,
           'Southern Ocean': so, 'Global': globe}

A = dict()
for r in regions:
    A[r] = 0.
```

```python
p.initialize_gradient_operator()
Nt = p.nc.variables['time'].shape[0]
for n in range(Nt):
    print n
    rho, Fheat, Fsalt, Fmix = p.dens_forcing[n]
    Fmix = np.ma.masked_invalid(Fmix)
    for r, reg in regions.iteritems():
        A[r] += reg.calculate_transformation_rate(rho, Fheat, Fsalt, Fmix)
for r in regions:
    A[r] /= Nt
```

```python
for F in (Fheat, Fsalt, Fmix):
    plt.figure()
    plt.imshow(F, cmap='RdBu_r')
    plt.clim(np.array([-1,1])*4e-5)
```

```python
Sv = 1e6
for regname, reg in regions.iteritems():
    plt.figure(figsize=(8,5))
    plt.plot(reg.rholevs[1:] - 1000, A[regname].T/Sv)
    plt.title(regname)
```
