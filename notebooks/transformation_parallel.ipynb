```python
from IPython.parallel import Client

c = Client()
dview = c.direct_view()
lview = c.load_balanced_view()

with dview.sync_imports():
    import numpy
    from watermasstools import pop_model, transformation
```

```python
ddir = '/glade/p/ncgd0001/hybrid_v5_rel04_BC5_ne120_t12_pop62/ocn-hist'
fprefix = 'hybrid_v5_rel04_BC5_ne120_t12_pop62.pop.h.nday1'
fnames = []
for year in xrange(47,86):
    for month in xrange(1,13):
        fnames.append(
         '%s/%s.%04d-%02d-01.nc' % (ddir, fprefix, year, month)
        )
```

```python
# load a test file
p = pop_model.POPFile(fnames[0])

# define basins
natl = transformation.WaterMassRegion(
                    basin_names=['Atlantic Ocean', 'GIN Seas', 'Labrador Sea'], latmin=15)
natl.initialize_mask(p)
natl.set_rholevs(numpy.load('../data/rholevs_natl.npy'))

npac = transformation.WaterMassRegion(
                    basin_names=['Pacific Ocean'], latmin=15)
npac.initialize_mask(p)
npac.set_rholevs(numpy.load('../data/rholevs_npac.npy'))

so = transformation.WaterMassRegion(
                    basin_names=['Southern Ocean'], latmax=30)
so.initialize_mask(p)
so.set_rholevs(numpy.load('../data/rholevs_so.npy'))

region_dict = {'natl': natl, 'npac': npac, 'so': so}
```

```python
# push to engines
dview.push(region_dict)
dview.execute("region_dict = {'natl': natl, 'npac': npac, 'so': so}")
# check
dview.execute('a = region_dict.keys()[0]')
a = dview.gather('a')
for r in a.get():
    assert r=='npac'
```

```python
def calc_transformation_rates(pop_fname):
    if isinstance(pop_fname, list):
        fname = pop_fname[0]
    else:
        fname = pop_fname
    
    print fname
    p = pop_model.POPFile(fname)
    p.initialize_gradient_operator()
    Nt = len(p.nc.variables['time'])
    
    result = dict()
    for rname in region_dict:
        result[rname] = 0.
    
    for n in range(Nt):
        print n
        F_temp, F_salt = p.dens_flux[n]
        
        H_ml = p.nc.variables['HMXL_2'][0]/100.
        rho0 = p.rho[n]
        bt = numpy.ma.masked_invalid(
            p.biharmonic_tendency(p.nc.variables['SST'][0]))
        bs = numpy.ma.masked_invalid(
            p.biharmonic_tendency(p.nc.variables['SSS'][0]))

        h0 = 10.
        D_salt = numpy.ma.masked_greater(
                    numpy.ma.masked_invalid(p.beta[n] * rho0 * bs), 1.)
        D_temp = numpy.ma.masked_greater(
                    numpy.ma.masked_invalid(-p.alpha[n] * rho0 * bt,), 1.)
        
        for rname, reg in region_dict.iteritems():
            A = reg.calculate_transformation_rate(rho0,
                              F_temp, F_salt, H_ml*D_temp, H_ml*D_salt, h0*D_temp, h0*D_salt)
            result[rname] += A

    for rname in region_dict:
        result[rname] /= Nt
        
    return result
```
