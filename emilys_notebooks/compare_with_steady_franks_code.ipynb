```python
import numpy as np
import scipy.io 
from matplotlib import pyplot as plt
%matplotlib inline

expts = ['lores_ctrl', 'hires_ctrl', 'lores_CO2', 'hires_CO2']
params = 'sigma2_hconst050'
Sv = 1e6
```

```python
MOCt = scipy.io.loadmat('/glade/scratch/enewsom/HR_ANALYSIS/PSI_SIGMA_147-167')
```

```python
MOC = MOCt['PSI_SIGMA']
```

```python
regions = ['natl', 'npac', 'so', 'globe']


A_trend = dict()
A_mean = dict()
rholevs = dict()
d = dict()
for expt in expts:
    A_trend[expt] = dict()
    A_mean[expt] = dict()
    rholevs[expt] = dict()
    for r in regions:
        d[r] = np.load('../data/WMT_%s_%s_%s.npz' % (expt, params, r))
        rholevs[expt][r] = d[r]['rholevs']
    T = np.arange(d['natl']['A'].shape[0])/12.

    for r in regions:
        # exclude mixing
        A_mean[expt][r] = d[r]['A'][:,:2].mean(axis=0)
        A_trend[expt][r] = ( d[r]['A'][-60:,:2].mean(axis=0)
                           - d[r]['A'][:60,:2].mean(axis=0) )
```

```python
F_sHR = -A_mean['hires_ctrl']['so'].sum(axis=0)/ Sv
F_sLR = -A_mean['lores_ctrl']['so'].sum(axis=0)/ Sv
```

```python
sigmaMOC = MOCt['s_moc']
latMOC = MOCt['lat_moc']
sigmaF = rholevs['hires_ctrl']['so'][1:] - 1000.
```

```python
plt.plot(sigmaMOC,-MOC[320,:],'r', linewidth=3)
plt.plot(sigmaF,F_sHR,'k', linewidth=3)
plt.legend(['MOC at -65','Net(hires)'], loc='lower left')
plt.xlim([36,37.9])
plt.xlabel(r'$\sigma_2$')
plt.ylabel('Sv')
plt.title('HiRes net transformation compared to isopycnal transport')
```
