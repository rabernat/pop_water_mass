## The Second Problem ##
You had an indentation error in this next cell, which caused the ```result[rname]``` line to be executed _outside_ of the loop. Also, you used the wrong key name (```rname``` instead of ```r```). Originally you had

    for r, reg in region_dict.iteritems():
         A= reg.calculate_transformation_rate(
                             rho,Fheat, Fsalt, Fmix,Fcab)
    result[rname] += A

But it should have been

    for r, reg in region_dict.iteritems():
         A= reg.calculate_transformation_rate(
                             rho,Fheat, Fsalt, Fmix,Fcab)
        result[r] += A

Remember that in python, indentation matters. I would also study up on how to [iterate over dictionaries](http://stackoverflow.com/questions/3294889/iterating-over-dictionaries-for-loops-in-python).

You can see below that the code now works.

````python
for n in range(Nt):
    print n
    rho, Fheat, Fsalt, Fmix, Fcab = p.dens_forcing[n]
    for r, reg in region_dict.iteritems():
        print r
        A = reg.calculate_transformation_rate(
                             rho,Fheat, Fsalt, Fmix,Fcab)
        result[r] += A
for r in region_dict:
    result[r] /= Nt
````
